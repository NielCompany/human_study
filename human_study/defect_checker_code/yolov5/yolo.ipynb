```python
import os
from IPython.display import Image, display

# ✅ 예측 명령어 구성
detect_cmd = """
python detect.py \
--weights runs/train/arrow_detect/weights/best.pt \
--source ../dataset/sum_images \
--conf 0.25 \
--save-txt \
--save-conf \
--project runs/detect \
--name arrow_result \
--exist-ok
"""

# ✅ 실행
print("🔍 YOLOv5 예측 중입니다...")
os.system(detect_cmd)

# ✅ 결과 이미지 1장 미리보기 (runs/detect/arrow_result 폴더 기준)
result_folder = "runs/detect/arrow_result"
files = os.listdir(result_folder)
image_files = [f for f in files if f.lower().endswith(('.jpg', '.png'))]

if image_files:
    print("✅ 예측된 이미지 예시:")
    display(Image(filename=os.path.join(result_folder, image_files[0])))
else:
    print("❌ 예측 이미지가 존재하지 않습니다.")
```
